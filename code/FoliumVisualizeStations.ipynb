```python
import pandas as pd
import folium as f
import geopandas as gp
```

## Load the data

```python
dati = gp.read_file("Postazioni 7/Postazioni.shp")
```

```python
dati.head()
```

```python
from pyproj import Transformer

# Define the source and target coordinate systems
source_crs = "EPSG:3857"  # Web Mercator projection (assumed)
target_crs = "EPSG:4326"  # WGS84 (latitude/longitude)

# Create a transformer object
transformer = Transformer.from_crs(source_crs, target_crs)

# Your coordinates
x, y = 1240351.431, 5523643.489

# Transform the coordinates
lon, lat = transformer.transform(x, y)

print(f"Latitude: {lat:.6f}, Longitude: {lon:.6f}")
```

## Add a popup as info for the user

```python
map = f.Map(location = [44.67265, 10.60645], zoom_start= 12)
for i, riga in dati.iterrows():
    x = riga["geometry"].coords.xy[0][0]
    y = riga["geometry"].coords.xy[1][0]

    # Create a transformer object
    transformer = Transformer.from_crs(source_crs, target_crs)

   

    # Transform the coordinates
    lon, lat = transformer.transform(x, y)
    
    print(lat, lon)
    nome_postazione = riga['POSTAZIONE']
    
    f.Marker(
      location=[lon, lat], popup= nome_postazione
      
   ).add_to(map)

#popup= nome_parco
```
