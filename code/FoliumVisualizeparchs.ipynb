```python
import pandas as pd
import folium as f
import geopandas as gp
```

## Load the data

```python
dati = gp.read_file("AMB_PARCHI_CARTELLI/AMB_PARCHI_CARTELLIPoint.shp")
```

```python
dati.head()
```

## Create a map

```python
m = f.Map()
m
```

## Center to location and add a zoom level

```python
m = f.Map(location = [44.67265, 10.60645], zoom_start= 12)
m
```

```python
map = f.Map(location = [44.67265, 10.60645], zoom_start= 12)
for i, riga in dati.iterrows():
    lat = riga["geometry"].coords.xy[0][0]
    lon = riga["geometry"].coords.xy[1][0]
    
    f.Marker(
      location=[lon, lat]
      
   ).add_to(map)
```

## Add the markers

```python
map
```

## Add a popup as info for the user

```python
map = f.Map(location = [44.67265, 10.60645], zoom_start= 12)
for i, riga in dati.iterrows():
    lat = riga["geometry"].coords.xy[0][0]
    lon = riga["geometry"].coords.xy[1][0]

    nome_parco = riga['PARCO']
    
    f.Marker(
      location=[lon, lat],popup= nome_parco
      
   ).add_to(map)
```
