```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
url= "https://en.wikipedia.org/wiki/List_of_action_films_of_the_2010s"
```

```python
dati_gz = pd.read_html(url)
```

```python
type(dati_gz)
```

```python
dati_gz[3]
```

```python
lista_df = []
count = 2010
for i in dati_gz: 

    if len(i) > 4:
        i['Anno'] = [count] * len(i)
        
        lista_df.append(i[1:])
        count = count +1
```

```python
dati = pd.concat(lista_df, ignore_index=True)
```

```python
dati_finali = dati.drop(["Subgenre/ notes", "Unnamed: 5", "Unnamed: 6", 0, 1, 2, 3, 4, 5, 6], axis = 1)
```

```python
filmPerPaese = dati_finali["Country"].value_counts().reset_index()
```

```python
paesiImportanti = filmPerPaese[0:20]
paesiImportanti
```

### Film per paese produttore

```python
plt.figure(figsize = (20, 5))
sns.barplot(data = paesiImportanti, x = paesiImportanti['Country'], y = paesiImportanti['count'])
plt.xticks(rotation = 45)
plt.show()
```

### Film per regista

```python
registi = dati_finali["Director"].value_counts().reset_index().head(20)
```

```python
plt.figure(figsize = (20, 5))
sns.barplot(data = registi, x = registi['Director'], y = registi['count'])
plt.xticks(rotation = 45)
plt.show()
```
