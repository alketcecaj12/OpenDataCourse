```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url1 = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)"
url2 = "https://en.wikipedia.org/wiki/List_of_countries_by_literacy_rate"
```

```python
gdp = pd.read_html(url1)
lr = pd.read_html(url2)
```

```python
data_gdp = gdp[2].iloc[:, 0:4]
data_gdp
```

```python
data_gdp.droplevel(0, axis = 1)
data_gdp.columns= ['Country', 'Forecast', 'Year', 'Estimate']
```

```python
data_lr = lr[1].droplevel(0, axis = 1)
```

```python
data_lr = data_lr.iloc[:, 0:2]
data_lr
```

```python
country_lr = []
for i, row in data_lr.iterrows(): 
    c = row['Country']
    print(c)
    c = c.strip().replace("*", "").strip()
    country_lr.append(c)
```

```python
country_gdp = []
for i, row in data_gdp.iterrows(): 
    c = row['Country']
    print(c)
    c = c.strip().replace("*", "")
    country_gdp.append(c)
```

```python
data_lr['Country'] = country_lr
data_gdp['Country'] = country_gdp
```

```python
data_gdp
```

```python
final_data= data_lr.merge(data_gdp, on ='Country' )
final_data.info()
```

```python
final_data = final_data.dropna()
```

```python
numbers = []
for i, row in final_data.iterrows(): 
    
    n = row['Rate']
    if "[" in n: 
        n = n.split('.')[0]
        numbers.append(n)
    else: 
        numbers.append(n)
```

```python
final_data['Rate'] = numbers
final_data['Rate'] = final_data['Rate'].astype(float) 
```

```python
final_data['Estimate'] = final_data['Estimate'].astype(float) 
```

```python
plt.figure(figsize = (15, 15))
plt.scatter(final_data['Rate'], final_data['Estimate'])
plt.show()
```
